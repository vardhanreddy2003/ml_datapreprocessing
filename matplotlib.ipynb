```python
x=np.linspace(start=-10,stop=10,num=200)
m=2
c=3
y=m*x+c
```

```python
plt.plot(x,y,label="y=2x+3")
plt.title("plot of linear equation")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(0,10)
plt.ylim(0,20)
plt.grid(True)
plt.legend()
plt.savefig("linear-equation.png")
plt.show()
```

```python
x=np.arange(10,100,step=1)
m=1
c=2
y=m*x+c
```

```python
plt.plot(x,y,label="y=x+2")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(0,20)
plt.ylim(0,20)
plt.show()
```

```python
x=np.linspace(-10,10,400)
a=2
b=3
c=1
y=a*x**2+b*x+c
plt.plot(x,y,label="y=2x^2+3x+1")
plt.title("plot of quadratic equation")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-10,10)
plt.ylim(0,100)
plt.grid(True)
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0.5,0.8,0.3,0.3])
plt.show()
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
fig=plt.figure(figsize=(8,6))
ax=fig.add_axes([0.1,0.1,0.85,0.85])
ax.plot(x,y,label="sin(x)",color="blue")
ax.set_title("Plot of Amplitude")
ax.set_xlabel("x")
ax.set_ylabel("Amplitude")
ax.legend()
plt.xlim(0,10)
plt.ylim(0,1)
plt.show()
```

```python
x=np.linspace(10,50,40)
y=np.cos(x)
fig=plt.figure(figsize=(6,5))
ax=fig.add_axes([0.1,0.1,0.85,0.85])
ax.plot(x,y,label="cos(x)",color="green")
ax.set_xlabel("X")
ax.set_ylabel("cos(x)")
plt.show()
```

```python
x=np.linspace(0,10,100)
fig=plt.figure(figsize=(8,10))

axes_center=fig.add_axes([0.3,0.3,0.4,0.4])#left,bottom,width,height
axes_center.plot(x,x+1,label="axes_center")
axes_center.set_title("axes_center")

axes_top=fig.add_axes([0.3,0.75,0.4,0.4])
axes_top.plot(x,2*x+1,label="axes_top")
axes_top.set_title("axes_top")

axes_top_right=fig.add_axes([0.75,0.75,0.4,0.4])
axes_top_right.plot(x,x-1)
axes_top_right.set_title("axes_top_right")

axes_left=fig.add_axes([0.05,0.3,0.2,0.4])
axes_left.plot(x,x*2)
axes_left.set_title("axes_left")
plt.show()
```
