```python
import numpy as np 
import pandas as  pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df=pd.read_csv("c:/Users/dell/Downloads/100-days-of-machine-learning-main/day36-imputing-numerical-data/titanic_toy.csv")
```

```python
df.head(3)
```

```python
df.info()
```

```python
df["Age_mean"]=df["Age"].fillna(df["Age"].mean())
df["Age_median"]=df["Age"].fillna(df["Age"].median())


df["Fare_mean"]=df["Fare"].fillna(df["Fare"].mean())
df["Fare_median"]=df["Fare"].fillna(df["Age"].median())
```

```python
df.head()
```

```python
print('Original Age variable variance: ', df['Age'].var())
print('Age Variance after median imputation: ', df['Age_median'].var())
print('Age Variance after mean imputation: ', df['Age_mean'].var())

print('Original Fare variable variance: ', df['Fare'].var())
print('Fare Variance after median imputation: ', df['Fare_median'].var())
print('Fare Variance after mean imputation: ', df['Fare_mean'].var())
```

```python
fig=plt.figure()

ax=fig.add_subplot(111)

df["Age"].plot(kind="kde",ax=ax,color="red")
df["Age_median"].plot(kind="kde",ax=ax,color="blue")
df["Age_mean"].plot(kind="kde",ax=ax,color="green")

lines,labels=ax.get_legend_handles_labels()

ax.legend(lines,labels)
```

```python
fig=plt.figure()

ax=fig.add_subplot()

df["Fare"].plot(kind="kde",ax=ax,color="red")
df["Fare_median"].plot(kind="kde",ax=ax,color="blue")
df["Fare_mean"].plot(kind="kde",ax=ax,color="green")

lines,labels=ax.get_legend_handles_labels()
ax.legend(lines,labels)
```
