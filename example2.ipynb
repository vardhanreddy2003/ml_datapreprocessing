```python
from  sklearn.compose import ColumnTransformer
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
transform=ColumnTransformer(transformers=[
    ("tnf1",OrdinalEncoder(categories=[["Poor","Average","Good"],["School","UG","PG"]]),[1,2]),
    ("tnf2",OneHotEncoder(sparse_output=False,drop="first"),[0])
],remainder="passthrough")
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns=["age","purchased"]),df["purchased"],test_size=0.2,random_state=0)
```

```python
X_train_transformed=transform.fit_transform(X_train)
```

```python
X_train_transformed=pd.DataFrame(X_train_transformed,columns=X_train.columns)
```
