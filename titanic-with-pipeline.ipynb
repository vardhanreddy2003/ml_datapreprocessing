```python
import numpy as np 
import pandas as pd
```

```python
df=pd.read_csv("c:/Users/dell/Downloads/100-days-of-machine-learning-main/day29-sklearn-pipelines/train.csv")
```

```python
df.head(2)
```

```python
df.drop(columns=["Name","Ticket","Cabin","PassengerId"],inplace=True)
```

```python
from sklearn.impute import SimpleImputer 
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.drop(labels="Survived",axis=1),df["Survived"],test_size=0.2,random_state=0)
```

```python
X_train
```

```python
X_train.iloc[:,[2]]
```

```python
trf1=ColumnTransformer([
    ("impute_age",SimpleImputer(),[2]),
    ("impute_embarked",SimpleImputer(strategy="most_frequent"),[6])
],remainder="passthrough")
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# Use sparse_output=False to ensure the output is a dense array
trf2 = ColumnTransformer([
    ('ohe_sex_embarked', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), [1, 6])
], remainder='passthrough')
```

```python
trf3=ColumnTransformer([("scale",MinMaxScaler(),slice(0,10))])
```

```python
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.tree import DecisionTreeClassifier
```

```python
trf4 = SelectKBest(score_func=chi2,k=8)
```

```python
trf5 = DecisionTreeClassifier()
```

```python

from sklearn.pipeline import Pipeline,make_pipeline

```

```python
pipe = Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3),
    ('trf4',trf4),
    ('trf5',trf5)
])
```

```python
X_train
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
import pickle
pickle.dump(pipe,open("pipe.pkl","wb"))
```
